### Hypothesis testing for average age

```python
#Set the hypothesis

#H0: mu age_new_process = age_old_process
#H1: mu age_new_process != age_old_process
```

```python
significance_level = 0.05
```

```python
age_new_process = clients_profile_df[clients_profile_df['Variation'] == 'Test']['client_age']
age_old_process = clients_profile_df[clients_profile_df['Variation'] == 'Control']['client_age']
age_new_process, age_old_process 
```

```python
st.ttest_ind(age_new_process ,age_old_process, equal_var=False)
```

```python
#the probability of observing the data given the null hypothesis is true is less than the significance level of 0.05.
#which suggests that there is a strong evidence against the null hypothesis
#thus, we reject the null hypothesis that the average ae for the test group is equal to the average age for the control group.
```

### Hypothesis Testing for Average Client Tenure

```python
#Set the hypothesis

#H0: mu years_new_process = years_old_process
#H1: mu years_new_process != years_old_process
```

```python
significance_level = 0.05
```

```python
years_new_process = clients_profile_df[clients_profile_df['Variation'] == 'Test']['client_years']
years_old_process = clients_profile_df[clients_profile_df['Variation'] == 'Control']['client_years']
years_new_process, years_old_process 
```

```python
st.ttest_ind(years_new_process ,years_old_process , equal_var=False)
```
